```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('project-2-hospital-data.csv')
df
```

### What is the average consultation fee for each department?


```python
df.groupby('Department')['Consultation_Fee'].mean()
```

### Which department sees the youngest patients on average?

```python
df.groupby('Department')['Patient_Age'].mean().sort_values()
```

### What is the trend of patient visits over the three months?


```python
daily_visit = df.groupby('Visit_Date').size()
daily_visit.plot()
plt.grid()
plt.xlabel('visit date')
plt.ylabel('Number of patients')
# daily_visit
```

### What is the gender distribution of patients?

```python
distribution = df['Gender'].value_counts()

distribution.plot(kind="bar")
plt.grid(axis='y')
```

### Which gender spends more on average consultation fees?

```python
df.groupby('Gender')['Consultation_Fee'].mean()
```

### What is the age distribution of patients in a histogram?

```python
df['Patient_Age'].plot(kind="hist", bins=10, edgecolor='black')
plt.xlabel('Age')
plt.tight_layout()
plt.show()
```

### Which department receives the most patients?

```python
dapartment_contribution = df['Department'].value_counts()

dapartment_contribution.plot(kind="bar")
plt.grid()
```

### How many follow-up visits were needed vs not?

```python
follow_up_data = df['Follow_Up'].value_counts()

follow_up_data.plot(kind='bar')
```
