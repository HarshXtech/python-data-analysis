#### Linear plot or line graph

##### It helps us to visualize continuous data or data that has linear relationship
##### it is usefull to identify trends and continous data
##### to graph linear plot in matplotlib, we use plt.plot() method

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(year, marks)
```

```python
x = [1,2,3,4,5,6]
y = [1, 4, 9, 16, 25, 36]
z = [1, 4, 6, 8, 10, 12]
# sns.set_style('darkgrid')
# plt.style.use('dark_background')
# two arguments!
plt.plot(x, y, marker="o", color='r', ls='--')
plt.plot(x, z, marker='D', color='b', ls=":")
plt.xlabel('Numbers')
plt.ylabel('square of the numbers')
plt.title('Int manipulation')
plt.grid(True)
plt.legend(["Square of the given numbers", "multiplication by 2"])
plt.show()
```

```python
y1 = []
y2 = []

x = range(-100, 100, 10)
```

```python
y1
```

```python
for i in x:
    y1.append(i**2)
```

```python
for i in x:
    y2.append(-i**2)
```

```python
y1
```

```python
y2
```

```python
plt.plot(x, y1, color='b')
plt.plot(x, y2, color='m')
plt.ylim(-2000, 2000)
plt.axhline(0)
plt.axvline(0)
plt.savefig('linear-plot.png')
```
