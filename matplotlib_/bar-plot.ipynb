### What is a Bar Plot?
###### A bar plot displays data in bars, either vertical or horizontal, to compare values across different categories.

###### Vertical bars are created with plt.bar()

###### Horizontal bars are created with plt.barh()

```python
import matplotlib.pyplot as plt
```

```python
names = ["Rahul", "Manthan", "Jay", "Saharsh", "Harsh"]
marks = [67, 10, 80, 90, 100]
```

```python
plt.bar(names, marks, color=['b', 'r', 'g'])
plt.plot(names, marks, color='r')
plt.grid(True)
```

```python
plt.barh(names, marks)
plt.grid(True)
plt.title('Comparison of students data')
plt.xlabel("Marks")
plt.ylabel("Student names")
```
