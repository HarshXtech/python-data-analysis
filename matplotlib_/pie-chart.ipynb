## What is Pie-chart?

### A pie chart is a circular statistical graphic that is divided into slices to illustrate numerical proportions. Each slice of the pie represents a categoryâ€™s contribution to the total. Pie charts are especially useful when you want to show how a whole is divided among different parts.

### Key Concepts of Pie Chart:
##### Circle: Represents the total (100%) of data.
##### Slices (or sectors): Represent individual data points or categories.

### Use Cases in Real Life:
##### Business: Show market share of companies.

##### Education: Visualize how students spend time on activities.

##### Environment: Breakdown of sources of pollution.

##### Finance: Show spending categories in a budget.

```python
category = ['Rent', 'Food', 'Transportation', 'Entertainment']
expenses = [7000, 3000, 2500, 1000]
explode = [0, 0, 0, 0.1]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.pie(expenses, labels=category, explode=explode, autopct="%.1f%%", colors=['r', 'g', 'b', 'm'], shadow=True)
plt.show()
```
