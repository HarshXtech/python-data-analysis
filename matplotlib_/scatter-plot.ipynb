#### A scatter plot is a type of data visualization that shows the relationship between two continuous variables using dots. Each dot represents an observation in your dataset, with its position determined by the values of the two variables.

### What is a Scatter Plot?
##### X-axis: one variable (independent)

##### Y-axis: another variable (dependent)

##### Each point (x, y): represents a single data observation

### Insights:

##### A scatter plot like this can show if there's a positive correlation between study hours and marks.

##### You could later fit a regression line if the relationship looks linear.

#### When to Use a Scatter Plot
##### To detect correlations or relationships between two variables.

##### To identify outliers or clusters in data.

##### Often used in regression analysis or machine learning as an initial data exploration step.
