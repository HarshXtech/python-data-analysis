## What is Histogram?

#### A histogram is a type of bar chart that represents the distribution of numerical data by grouping values into intervals (called bins). In Python, histograms are commonly created using libraries like Matplotlib and Seaborn.

#### Histograms are useful for visualizing the frequency distribution, identifying patterns, and detecting outliers in data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# standard normal distribution
data = np.random.randn(1000)
```

```python
data
```

```python
plt.hist(data, histtype='bar', bins=5)
```

```python
x = np.random.randint(0, 100, 15)
```

```python
x
```

```python
rng = (0, 100)
plt.hist(x, range=rng, color='g', rwidth=0.5, alpha=0.5)
```

```python
import seaborn as sns
```

```python
sns.histplot(data, kde=True)
```
